```python
from fastai.vision.all import *

image_path = Path(input("Enter the path to the image you want to predict: "))

data = ImageDataLoaders.from_folder('/content/seg_train',
                                     valid_pct=0.2,
                                     item_tfms=Resize(224),
                                     batch_tfms=aug_transforms(),
                                     num_workers=4)

learn = cnn_learner(data, models.resnet34, metrics=accuracy)

img = PILImage.create(image_path)
pred_class, pred_idx, outputs = learn.predict(img)

print(f'Predicted Class: {pred_class}, Index: {pred_idx}, Outputs: {outputs}')
```
